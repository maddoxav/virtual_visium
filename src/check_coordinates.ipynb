```python
from tiler_the_creator import *
from openslide import open_slide
from pathlib import Path
import numpy as np
import feather
import matplotlib.pyplot as plt
```

```python
# checking that tissue coordinates align with wsi
sample_path = Path(r"C:\Users\maddoxav\OneDrive - Michigan Medicine\Documents\Rao_Lab\data\GBM_spatial_transcriptomics\Sample-1512")
slide_fname = "20H214_Al-Holou_GEX_Slide_Square_A.tif" 
slide_path = sample_path / "TP_Images" / slide_fname 
slide = open_slide(str(slide_path))

dims = slide.dimensions
scale_by = 40
scaled_dims = tuple(x / scale_by for x in dims)

thumb = slide.get_thumbnail(scaled_dims)
thumb_arr = np.array(thumb)

tc_fname = "brain1A-s1512-tc.feather"
tc_path = sample_path / "assay_data" / tc_fname 
tc = feather.read_dataframe(tc_path)

fig, ax = plt.subplots()
x = tc['imagecol'] / scale_by
y = tc['imagerow'] / scale_by
ax.imshow(thumb_arr)
ax.scatter(x, y, s=1)
```
